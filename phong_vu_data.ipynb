```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://phongvu.vn/laptop-va-linh-kien-macbook-715.cat?pv_medium=m-laptop-va-linh-kien-macbook&page=1'
html_text = requests.get(url).text
tree = BeautifulSoup(html_text, 'html.parser')
all_products = tree.find_all('a', class_= 'css-1rhapru')

for product in all_products:
    if product.find('div', class_= 'css-1qgxba5') != None:
        product_name = product.find('div', class_= 'css-1qgxba5').string.strip()
    else:
        product_name = product.find('div', class_= 'css-154c1mb').string.strip()

    if product.find('div', class_= 'css-l9dhqu') != None:
        product_price_af = product.find('div', class_= 'css-l9dhqu').string.strip()
        product_price_be = None
    else:
        product_price_af = product.find('span', class_= 'css-1ou7728').text
        if product.find('span', class_= 'css-nq9pgm') != None:
            product_price_be = product.find('span', class_= 'css-nq9pgm').text
        else:
            product_price_be = None
    image_url = product.find('img', class_= 'lazyload css-jdz5ak')['src']
    print(product_name)
    print(image_url)
    print('\n')
```

```python
number_laptop_pages = 18 # number of Laptop and accessories page that i had found out in reality by view their page source
laptop_out = open('laptop_out.csv', 'w', encoding = 'utf-8')
laptop_out.write('Tên sản phẩm,Giá trước khi giảm,Giá sau khi giảm,Image_url\n')
count = 0
for n in range(number_laptop_pages + 1):
    print(count); count += 1
    url = f'https://phongvu.vn/laptop-va-linh-kien-macbook-715.cat?pv_medium=m-laptop-va-linh-kien-macbook&page={n}'
    html_text = requests.get(url).text
    tree = BeautifulSoup(html_text, 'html.parser')
    all_products = tree.find_all('a', class_= 'css-1rhapru')
    for product in all_products:
        if product.find('div', class_= 'css-1qgxba5') != None:
            product_name = product.find('div', class_= 'css-1qgxba5').string.strip()
        else:
            product_name = product.find('div', class_= 'css-154c1mb').string.strip()

        if product.find('div', class_= 'css-l9dhqu') != None:
            product_price_af = product.find('div', class_= 'css-l9dhqu').string.strip()
            product_price_be = None
        else:
            product_price_af = product.find('span', class_= 'css-1ou7728').text.strip()
            if product.find('span', class_= 'css-nq9pgm') != None:
                product_price_be = product.find('span', class_= 'css-nq9pgm').text.strip()
            else:
                product_price_be = None
        image_url = product.find('img', class_= 'lazyload css-jdz5ak')['src']
        print(f'{product_name}\t{product_price_be}\t{product_price_af}\n')
        laptop_out.write(f'{product_name},{product_price_be},{product_price_af},{image_url}\n')
        time.sleep(3)
laptop_out.close()
print('Done')
```
