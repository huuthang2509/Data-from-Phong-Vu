```python
url = 'https://phongvu.vn/laptop-va-linh-kien-macbook-715.cat?pv_medium=m-laptop-va-linh-kien-macbook&page=1'
html_text = requests.get(url).text
tree = BeautifulSoup(html_text, 'html.parser')
all_products = tree.find_all('a', class_= 'css-1rhapru')

for product in all_products:
    if product.find('div', class_= 'css-1qgxba5') != None:
        product_name = product.find('div', class_= 'css-1qgxba5').string.strip()
    else:
        product_name = product.find('div', class_= 'css-154c1mb').string.strip()

    if product.find('div', class_= 'css-l9dhqu') != None:
        product_price_af = product.find('div', class_= 'css-l9dhqu').string.strip()
        product_price_be = None
    else:
        product_price_af = product.find('span', class_= 'css-1ou7728').text
        if product.find('span', class_= 'css-nq9pgm') != None:
            product_price_be = product.find('span', class_= 'css-nq9pgm').text
        else:
            product_price_be = None
    image_url = product.find('img', class_= 'lazyload css-jdz5ak')['src'] #product in list page

    # Detail page
    url_temp = 'https://phongvu.vn' + product['href']
    html_text_temp = requests.get(url_temp).text
    tree_temp = BeautifulSoup(html_text_temp, 'html.parser')
    all_image = tree_temp.find('div', class_= 'css-bkflzc').find_all('img')
    # all_image[0]['src']
    all_image_str = ''
    for image in all_image:
        all_image_str += image['src'] + '\n'
    product_ths = tree_temp.find('div', {'class': 'css-1bglpso'}).find_all('div', {'class': 'css-7j9rw7'}) # table thong so ki thuat
    product_tskt = '' # chuyen ve dang str
    for item in product_ths:
        product_tskt += item.find('span', {'class': 'css-6z2lgz'}).text + ': ' + item.find('div', {'class': 'css-111s35w'}).text + '; '
    product_tskt = product_tskt[:-1]
    product_detail = tree_temp.find('div', {'class': 'css-fhqexl'}).find('div', {'class': 'css-111s35w'})
    print(product_name)
    print(image_url)
    print(product_price_be)
    print(product_price_af)
    print(all_image_str)
    print(product_tskt)
    print(product_detail)
    print('\n')
    time.sleep(3)
```
